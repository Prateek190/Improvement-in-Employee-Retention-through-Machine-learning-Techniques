```
import pandas as pd
```

```
url='https://github.com/Prateek190/data/raw/master/resource_consumption(x_i).csv'
data1=pd.read_csv(url)
```

```
# Employees corresponding resource consumption_1 histogram
import matplotlib.pyplot as plt

import seaborn as sns


plt.title(" Employees corresponding resource consumption_1")
plt.xlabel("resource consumption_1")
plt.ylabel("No. of  Employees")
sns.distplot(data1,bins=70,kde=False,rug=False)
plt.savefig("Employees corresponding resource consumption_1dist")
```

```
url='https://github.com/Prateek190/data/raw/master/r1%20vs%20r2.csv'
data=pd.read_csv(url)
```

```
 # resource consumption_1 vs  resource consumption_2: years vs no. of employees line plot


sns.kdeplot(data['r1'],shade=True,color='r')
sns.kdeplot(data['r2'],shade=True,color='b')
plt.xlabel('resource consumption')
plt.ylabel("frequency")
plt.title(" resource consumption_1 vs  resource consumption_2")

plt.savefig(" resource consumption_1 vs  resource consumption_2")
```

```
url='https://github.com/Prateek190/data/raw/master/new.csv'
data3=pd.read_csv(url)
```

```
#resource consumption_2 : employee vs resource_2 line plot

sns.set(style="darkgrid")
sns.kdeplot(data3['Employee'],data3['resource consumption_2'],shade=True,color='r')
#sns.relplot(x=data3['e'], y=data3['r1'],ci=None,size=1, size_norm=(0,1200),kind="line" , data=data3)
plt.title("resource consumption_2 for each Employee")
plt.savefig(' resource consumption_2: employee vs yearsline')
```

```
sns.set(style="darkgrid")
sns.relplot(x="Employee", y="resource consumption_1", data=data3);
plt.title("resource consumption_1 for each Employee")
plt.savefig("resource consumption_1 for each Employee")
```

```
sns.set(style="darkgrid")
sns.relplot(x="Employee", y="resource consumption_2", data=data3);
plt.title("resource consumption_2 for each Employee")
#plt.savefig("Expected tenure_2 for each Employee")
```
