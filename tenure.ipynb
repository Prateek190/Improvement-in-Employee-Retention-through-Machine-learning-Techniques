```
import pandas as pd
```

```
url='https://github.com/Prateek190/data/raw/master/tenure1_final.csv'
data1= pd.read_csv(url)
```

```
# Expected tenure_1: years vs no. of employees histogram
import matplotlib.pyplot as plt

import seaborn as sns


plt.title("No. of Employees corresponding to expected tenure_1")
plt.xlabel("years")
plt.ylabel("No. of  Employees")
sns.distplot(data1,bins=70,kde=False,rug=False)
plt.savefig("Expected tenure_1: years vs no")
```

```
url='https://github.com/Prateek190/data/raw/master/Book1.csv'
data=pd.read_csv(url)
```

```
 #Expected tenure_1 vs Expected tenure_2: years vs no. of employees line plot


sns.kdeplot(data['Expected Tenure 1'],shade=True,color='r')
sns.kdeplot(data['Expected Tenure 2'],shade=True,color='b')
plt.xlabel('Years')
plt.ylabel("frequency")
plt.title("Expected tenure_1 vs Expected tenure_2")
#plt.show()
plt.savefig("Expected tenure_1 vs Expected tenure_2")
```
