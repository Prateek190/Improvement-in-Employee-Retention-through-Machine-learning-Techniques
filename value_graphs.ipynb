```
import pandas as pd
```

```
url='https://github.com/Prateek190/data/raw/master/value_predictor(x_i).csv'
data=pd.read_csv(url)
```

```
# Employees corresponding value prediction_1 histogram
import matplotlib.pyplot as plt

import seaborn as sns


plt.title(" Employees corresponding value prediction_1 ")
plt.xlabel("value prediction_1")
plt.ylabel("No. of  Employees")
sns.distplot(data['value predictor(x_i)'],bins=70,kde=False,rug=False)
plt.savefig("Employees corresponding resource value prediction_1dist")
```

```
# Employees corresponding value prediction_2 histogram
import matplotlib.pyplot as plt

import seaborn as sns


plt.title(" Employees corresponding value prediction_2 ")
plt.xlabel("value prediction_2")
plt.ylabel("No. of  Employees")
sns.distplot(data['value predictor(x_change)'],bins=70,kde=False,rug=False)
plt.savefig("Employees corresponding resource value prediction_2dist")
```

```
 # value prediction_1 vs value prediction_2: years vs no. of employees line plot


sns.kdeplot(data['value predictor(x_i)'],shade=True,color='r')
sns.kdeplot(data['value predictor(x_change)'],shade=True,color='b')
plt.xlabel('value prediction')
plt.ylabel("frequency")
plt.title(" value prediction_1 vs value prediction_2 ")

plt.savefig("  value prediction_1 vs value prediction_2")
```

```
# Employees corresponding def_value1 histogram
import matplotlib.pyplot as plt

import seaborn as sns


plt.title(" Employees corresponding def_value1 ")
plt.xlabel("def_value1")
plt.ylabel("No. of  Employees")
sns.distplot(data['def_value1'],bins=70,kde=False,rug=False)
plt.savefig("Employees corresponding resource def_value1dist")
```

```
# Employees corresponding def_value2 histogram
import matplotlib.pyplot as plt

import seaborn as sns


plt.title(" Employees corresponding def_value2 ")
plt.xlabel("def_value2")
plt.ylabel("No. of  Employees")
sns.distplot(data['def_value2'],bins=70,kde=False,rug=False)
plt.savefig("Employees corresponding resource def_value2dist")
```

```
 # def_value1 vs  def_value2 : years vs no. of employees line plot


sns.kdeplot(data['def_value1'],shade=True,color='r')
sns.kdeplot(data['def_value2'],shade=True,color='b')
plt.xlabel('def_value')
plt.ylabel("frequency")
plt.title("def_value1 vs  def_value2")

plt.savefig("def_value1 vs  def_value2")
```
